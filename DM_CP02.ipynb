Classification : Severity of illness (on basis of health service area, gender, and diagnosis).

```python
#'APR Severity of Illness Code', 'APR Severity of Illness Description'
df['APR Severity of Illness Description'].unique()
```

The null values in severity of illness and the selected independent variables is < 1%. Hence, we can drop the null values.

```python
df1 = df[['Health Service Area', 'Gender', 'CCS Diagnosis Code', 'APR Severity of Illness Description']].dropna()
df1.columns = ['area', 'gender', 'diagnosis', 'severity_of_illness']
df1.head()
```

```python
df1['gender'].replace(to_replace='U', value='F', inplace=True)
```

```python
df1.describe(include='object')
```

```python
#Print percentage of value counts
for col in df1.columns:
  print('Column:', col)
  print(100 * df1[col].value_counts() / len(df1))
  print('-'*40)
```

```python
#applying one-hot encoding to area, gender, race and ethnicity columns
df_encoded = pd.get_dummies(df1, columns=['area', 'gender'])
```

```python
df_encoded.head()
```

```python
#ordinal encoding severity of illness
custom_mapping = {
    'Minor': 1,
    'Moderate': 2,
    'Major': 3,
    'Extreme':4
}

# Use the map() function to apply the custom mapping
df_encoded['severity'] = df_encoded['severity_of_illness'].map(custom_mapping)

# Display the DataFrame with the custom ordinal encoding
df_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical column to encode the labels
df_encoded['diagnosis_encoded'] = label_encoder.fit_transform(df_encoded['diagnosis'])
```

```python
df_final = df_encoded.drop(['severity_of_illness'], axis=1)
df_final.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_final.drop(['severity', 'diagnosis'], axis=1)
y = df_final['severity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
```

We shall use ordinal logistic regression, logistic regression, and random forest classifier

```python
#!pip install mord
```

Ordinal Logistic Regression

```python
from mord import LogisticAT

model = LogisticAT(alpha=0.01)
model.fit(X_train[:10000], y_train[:10000])
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, zero_division=0))
```

Let us perform search for hyperparameter alpha:

```python
accs = []
for alpha in [0, 0.001, 0.01, 1]:
  model = LogisticAT(alpha=alpha)
  model.fit(X_train[:10000], y_train[:10000])

  y_pred = model.predict(X_train[10000:10500])
  y_true = y_train[10000:10500]

  acc = (y_true==y_pred).astype(int).sum() / 500
  accs.append(acc)

accs
```

As we can see, none of the combinations produces good accuracy


Let us try logistic regression

```python
from sklearn.linear_model import LogisticRegression

accs = []
for alpha in [0.001, 0.01, 1, 10]:
  model = LogisticRegression(penalty='l2', C=alpha)
  model.fit(X_train[:10000], y_train[:10000])

  y_pred = model.predict(X_train[10000:10500])
  y_true = y_train[10000:10500]

  acc = (y_true==y_pred).astype(int).sum() / 500
  accs.append(acc)

accs
```

Logistic regression with C=0.001 produces the best accuracy. Let us train the model with this parameter and more data.

```python
#Taking 1 Lakh values
model = LogisticRegression(penalty='l2', C=0.001)
model.fit(X_train[:100000], y_train[:100000])

y_pred = model.predict(X_test)

# Print a classification report with precision, recall, and F1-score
report = classification_report(y_test, y_pred)
print(report)
```

We get an accuracy of 40%

Finally, let us try Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the training data
rf_classifier.fit(X_train[:10000], y_train[:10000])

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Print a classification report with precision, recall, and F1-score
report = classification_report(y_test, y_pred)
print(report)
```

Let us perform search over hyperparameter n_estimators:

```python
from sklearn.linear_model import LogisticRegression

accs = []
for n in [25, 50, 75, 100, 200]:
  rf_classifier = RandomForestClassifier(n_estimators=n, random_state=42)

  # Fit the model on the training data
  rf_classifier.fit(X_train[:10000], y_train[:10000])

  # Make predictions on the test data
  y_pred = rf_classifier.predict(X_train[10000:10500])
  y_true = y_train[10000:10500]

  acc = (y_true==y_pred).astype(int).sum() / 500
  accs.append(acc)

accs
```

Best accuracy is produced by n_estimators = 75. Let us train the model with more data.

```python
rf_classifier = RandomForestClassifier(n_estimators=75, random_state=42)

rf_classifier.fit(X_train[:100000], y_train[:100000])
```

```python
y_pred = rf_classifier.predict(X_test)
print(classification_report(y_test, y_pred, zero_division=0))
```

We get an accuracy of 51% on the test set.
