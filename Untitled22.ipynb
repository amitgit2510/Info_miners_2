```python
df2=df.copy()
```

```python
columns_to_check = ['Health Service Area', 'Hospital County', 'Operating Certificate Number','Facility ID','Zip Code - 3 digits','Patient Disposition','CCS Diagnosis Code','CCS Diagnosis Description','CCS Procedure Code','CCS Procedure Description','APR Severity of Illness Description','APR Risk of Mortality']  # Replace with your desired column names

df2=df2.dropna(subset=columns_to_check)
```

```python
relevant_columns = ['Health Service Area', 'Operating Certificate Number', 'Facility ID', 'Age Group', 'Gender', 'Type of Admission', 'Patient Disposition', 'APR Medical Surgical Description', 'APR Severity of Illness Code', 'APR Risk of Mortality', 'Total Charges']

dfe1 = dfe[relevant_columns]
```

```python
def map_charges(value):
    if value < 10000:
        return 0
    elif 10000 <= value < 30000:
        return 1
    elif 30000 <= value < 60000:
        return 2
    elif 60000 <= value < 100000:
        return 4
    else:
        return 5

dfe1['Bill'] = dfe1['Total Charges'].apply(map_charges)

print(dfe1)
```

```python
dfe1.head()
```

```python
X = dfe1.drop(['Bill'], axis=1)

y = dfe1['Bill']
```

```python
# split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier


rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
from sklearn.metrics import accuracy_score

print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", linewidths=0.5)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy score: {accuracy:.4f}")
```
